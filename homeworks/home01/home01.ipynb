```python
def rgb2grey(image):
    image_g = []
    for i in range(0,image.shape[0]):
        for j in range(0,image.shape[1]):
            print(i,j,image[i,j,0])
            image_g[i,j] = (image[i,j,0] + image[i,j,1] + image[i,j,2]) / 3
    return image_g
```
