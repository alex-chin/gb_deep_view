### Цветовая обработка: преобразование из цветного RGB в GRAYSCALE

> Реализация на чистом Python и списках. Оставлено только чтение исходного файла

```python
def arr_init(m, n):
    return [[0] * n for i in range(m)]


def rgb2grey(image):
    m = image.shape[0]
    n = image.shape[1]
    image_g = arr_init(n, m)
    for j in range(0, m):
        for i in range(0, n):
            image_g[i][j] = (image[j, i, 0] + image[j, i, 1] + image[j, i, 2]) / 3
    return splash(image_g)


def splash(image):
    m = len(image)
    n = len(image[0])
    image_new = arr_init(n, m)
    for i in range(m):
        for j in range(n):
            image_new[j][i] = image[i][j]
    return image_new
```

```python
img_grey = rgb2grey(img)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img_grey, cmap='gray')
```

### Проверка преобразований


```python
from skimage.color import rgb2hsv, hsv2rgb
```

```python
img_hsv = rgb2hsv(img)
img_hsv_new = np.copy(img_hsv)
img_hsv_new[:, :, 1] /= 3
img_new = hsv2rgb(img_hsv_new)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img_new, cmap='hsv')
```
